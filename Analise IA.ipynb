```python

# Importar bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from transformers import pipeline
from google.colab import files

# Upload do arquivo Excel
print("Selecione seu arquivo Excel (ex: Ecommerce Novo.xlsx):")
uploaded = files.upload()
arquivo_excel = list(uploaded.keys())[0]

# Ler todas as abas
xls = pd.ExcelFile(arquivo_excel)
print("Abas disponíveis:", xls.sheet_names)

# Criar pipeline IA leve
nlp = pipeline("text-generation", model="EleutherAI/gpt-neo-125M")

# Processar cada aba
for aba in xls.sheet_names:
    print("\n" + "="*40)
    print(f"Aba: {aba}")
    print("="*40)

    df = pd.read_excel(xls, sheet_name=aba)
    df = df.rename(columns=lambda x: x.strip().lower().replace(" ", "_"))

    if 'data_venda' not in df.columns or 'receita_real' not in df.columns or 'canal' not in df.columns:
        print("Aba não contém colunas essenciais ('data_venda', 'receita_real', 'canal'), pulando...")
        continue

    df['data_venda'] = pd.to_datetime(df['data_venda'])
    df['mes'] = df['data_venda'].dt.to_period('M')

    # Indicadores
    faturamento_total = df['receita_real'].sum()
    ticket_medio = df['receita_real'].mean()
    vendas_mes = df.groupby('mes')['receita_real'].sum()
    vendas_canal = df.groupby('canal')['receita_real'].sum()

    # Converter para string para IA
    vendas_mes_str = {str(k): v for k, v in vendas_mes.round(2).to_dict().items()}
    vendas_canal_str = {str(k): v for k, v in vendas_canal.round(2).to_dict().items()}

    # Texto legível
    vendas_mes_texto = "\n".join([f"{k}: R$ {v:,.2f}" for k, v in vendas_mes_str.items()])
    vendas_canal_texto = "\n".join([f"{k}: R$ {v:,.2f}" for k, v in vendas_canal_str.items()])

    resumo_texto = f"""
Faturamento Total: R$ {faturamento_total:,.2f}
Ticket Médio: R$ {ticket_medio:,.2f}
Faturamento por Mês:
{vendas_mes_texto}
Faturamento por Canal:
{vendas_canal_texto}
"""
    # Gráficos
    plt.figure(figsize=(10,5))
    sns.barplot(x=list(vendas_mes_str.keys()), y=list(vendas_mes_str.values()))
    plt.xticks(rotation=45)
    plt.title(f"Faturamento por Mês - {aba}")
    plt.ylabel("R$")
    plt.show()

    plt.figure(figsize=(6,4))
    sns.barplot(x=list(vendas_canal_str.keys()), y=list(vendas_canal_str.values()))
    plt.title(f"Faturamento por Canal - {aba}")
    plt.ylabel("R$")
    plt.show()

    # Prompt IA
    prompt = f"""
Você é um consultor de BI. Analise estes dados resumidos e gere 3 insights estratégicos claros sobre o negócio.
- Seja objetivo e curto
- Não repita números
- Foque em estratégia e decisões de negócio

Dados:
{resumo_texto}
"""
    resposta = nlp(prompt, max_new_tokens=200, do_sample=True, temperature=0.7)

    print("\n=== INSIGHTS DA IA - ABA: {} ===\n".format(aba))
    print(resposta[0]['generated_text'])
    print("\n=====================\n")
```

```python

# MINI DASHBOARD DE BI - E-COMMERCE

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Upload do Excel
# uploaded = files.upload()
# arquivo_excel = list(uploaded.keys())[0]

# Ler Excel
xls = pd.ExcelFile(arquivo_excel)
df = pd.read_excel(xls, sheet_name=xls.sheet_names[0])

# Padronizar colunas
df = df.rename(columns=lambda x: x.strip().lower().replace(" ", "_"))
df['data_venda'] = pd.to_datetime(df['data_venda'])
df['mes'] = df['data_venda'].dt.to_period('M')

# Indicadores principais
faturamento_total = df['receita_real'].sum()
ticket_medio = df['receita_real'].mean()
vendas_mes = df.groupby('mes')['receita_real'].sum()
vendas_canal = df.groupby('canal')['receita_real'].sum()

# Gráfico: Faturamento por Mês
plt.figure(figsize=(10,5))
sns.barplot(x=[str(m) for m in vendas_mes.index], y=vendas_mes.values)
plt.title("Faturamento por Mês")
plt.ylabel("R$")
plt.xlabel("Mês")
plt.xticks(rotation=45)
plt.show()

# Gráfico: Faturamento por Canal
plt.figure(figsize=(6,4))
sns.barplot(x=vendas_canal.index, y=vendas_canal.values)
plt.title("Faturamento por Canal")
plt.ylabel("R$")
plt.show()

# Exibir indicadores resumidos
print("=== INDICADORES PRINCIPAIS ===")
print(f"Faturamento Total: R$ {faturamento_total:,.2f}")
print(f"Ticket Médio: R$ {ticket_medio:,.2f}\n")

# Insights estratégicos (do relatório anterior)
insights = [
    "1️⃣ Marketplace domina 51% do faturamento. Investir em marketing e logística neste canal.",
    "2️⃣ Picos em setembro e janeiro, queda em março. Planejar estoques e promoções por mês.",
    "3️⃣ Ticket médio elevado indica produtos de maior valor. Usar cross-sell e upsell para aumentar volume."
]

print("=== INSIGHTS ESTRATÉGICOS ===")
for insight in insights:
    print(insight)
```
